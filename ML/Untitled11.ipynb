```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine_quality = fetch_ucirepo(id=186)

# data (as pandas dataframes)
X = wine_quality.data.features
y = wine_quality.data.targets

# metadata
print(wine_quality.metadata)

# variable information
print(wine_quality.variables)
```

```python
# =========================
# Notebook: Analyse & k-NN
# Dataset: Wine Quality (white)
# Collez ce code dans Google Colab (séparez en cellules si vous voulez)
# =========================

# 0) Imports + paramètres
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize'] = (10,6)

# 1) Chargement des données
link = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
df = pd.read_csv(link, delimiter=';')
print("Shape (n_samples, n_features+1):", df.shape)
print("\nColonnes:\n", df.columns.tolist())
print("\nQuelques premières lignes:")
display(df.head())

# 2) Analyse descriptive rapide
print("\n=== Description statistique (numerical) ===")
display(df.describe().T)

print("\n=== Comptage des classes (quality) ===")
display(df['quality'].value_counts().sort_index())

# transformer en problème binaire: 0 = "bad" (<=5), 1 = "good" (>=6)
df['quality_bin'] = (df['quality'] > 5).astype(int)
print("\nRépartition binaire (0=bad,1=good):")
display(df['quality_bin'].value_counts())

# 3) Visualisations descriptives
# boxplots des features
features = df.columns.drop(['quality','quality_bin'])
plt.figure(figsize=(12,6))
sns.boxplot(data=df[features], orient='v')
plt.xticks(rotation=90)
plt.title("Boxplots des caractéristiques physico-chimiques")
plt.tight_layout()
plt.show()

# heatmap corrélation
plt.figure(figsize=(10,8))
corr = df[features].corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title("Matrice de corrélation (features)")
plt.show()

# 4) Préparation X, Y et split train/val/test (stratifié)
X = df[features].copy()
Y = df['quality_bin'].copy()

# split initial: test_size = 1/3 puis split remaining into train/val equally
Xa, Xt, Ya, Yt = train_test_split(X, Y, test_size=1/3, shuffle=True, stratify=Y, random_state=42)
Xa, Xv, Ya, Yv = train_test_split(Xa, Ya, test_size=0.5, shuffle=True, stratify=Ya, random_state=42)

print("\nTailles: train={}, val={}, test={}".format(Xa.shape[0], Xv.shape[0], Xt.shape[0]))
print("Proportions y (train/val/test):",
      Ya.value_counts(normalize=True).to_dict(),
      Xv.shape[0], Yv.value_counts(normalize=True).to_dict(),
      Xt.shape[0], Yt.value_counts(normalize=True).to_dict()
     )

# 5) k-NN sur données brutes (non normalisées)
k_vector = np.arange(1, 40, 2)  # 1,3,5,...,39
error_train = np.empty(len(k_vector))
error_val = np.empty(len(k_vector))

for idx, k in enumerate(k_vector):
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(Xa, Ya)
    ypred_train = clf.predict(Xa)
    ypred_val = clf.predict(Xv)
    error_train[idx] = 1 - accuracy_score(Ya, ypred_train)
    error_val[idx] = 1 - accuracy_score(Yv, ypred_val)

# Plot erreurs
plt.figure()
plt.plot(k_vector, error_train, marker='o', label='Erreur train (non norm)')
plt.plot(k_vector, error_val, marker='o', label='Erreur val (non norm)')
plt.xlabel('k (n_neighbors)')
plt.ylabel('Error rate')
plt.title('k-NN error vs k (non normalisé)')
plt.legend()
plt.grid(True)
plt.show()

# Sélection du meilleur k d'après validation (non normalisé)
ind_opt = np.argmin(error_val)
k_star = k_vector[ind_opt]
print(f"Meilleur k (non normalisé) selon validation: k* = {k_star} avec erreur val = {error_val[ind_opt]:.4f}")

# Évaluer sur test avec k_star (non normalisé)
clf_star = KNeighborsClassifier(n_neighbors=k_star)
clf_star.fit(Xa, Ya)
ypred_test = clf_star.predict(Xt)
error_test = 1 - accuracy_score(Yt, ypred_test)
print(f"Erreur test (non normalisé) pour k*={k_star}: {error_test:.4f}")
print("\nConfusion matrix (test):")
display(pd.DataFrame(confusion_matrix(Yt, ypred_test),
                     index=['true_0','true_1'], columns=['pred_0','pred_1']))
print("\nClassification report (test):")
print(classification_report(Yt, ypred_test))

# 6) Normalisation (StandardScaler) — fit sur Xa seulement, appliquer à val & test
sc = StandardScaler(with_mean=True, with_std=True)
sc.fit(Xa)               # Important: on fit uniquement sur l'ensemble d'entraînement
Xa_n = sc.transform(Xa)
Xv_n = sc.transform(Xv)
Xt_n = sc.transform(Xt)

# 7) Répéter k-NN sur données normalisées
error_train_n = np.empty(len(k_vector))
error_val_n = np.empty(len(k_vector))

for idx, k in enumerate(k_vector):
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(Xa_n, Ya)
    ypred_train = clf.predict(Xa_n)
    ypred_val = clf.predict(Xv_n)
    error_train_n[idx] = 1 - accuracy_score(Ya, ypred_train)
    error_val_n[idx] = 1 - accuracy_score(Yv, ypred_val)

# Plot comparaison avant/après normalisation (validation)
plt.figure()
plt.plot(k_vector, error_val, marker='o', label='val (non norm)')
plt.plot(k_vector, error_val_n, marker='o', label='val (normalisé)')
plt.xlabel('k')
plt.ylabel('Error rate (validation)')
plt.title('Comparaison validation error: non-norm vs normalized')
plt.legend()
plt.grid(True)
plt.show()

# Afficher erreurs train/val normalisé
plt.figure()
plt.plot(k_vector, error_train_n, marker='o', label='train (norm)')
plt.plot(k_vector, error_val_n, marker='o', label='val (norm)')
plt.xlabel('k')
plt.ylabel('Error rate')
plt.title('k-NN error vs k (normalisé)')
plt.legend()
plt.grid(True)
plt.show()

# Meilleur k après normalisation
ind_opt_n = np.argmin(error_val_n)
k_star_n = k_vector[ind_opt_n]
print(f"Meilleur k (normalisé) selon validation: k* = {k_star_n} avec erreur val = {error_val_n[ind_opt_n]:.4f}")

# Évaluer sur test avec k_star_n (normalisé)
clf_star_n = KNeighborsClassifier(n_neighbors=k_star_n)
clf_star_n.fit(Xa_n, Ya)
ypred_test_n = clf_star_n.predict(Xt_n)
error_test_n = 1 - accuracy_score(Yt, ypred_test_n)
print(f"Erreur test (normalisé) pour k*={k_star_n}: {error_test_n:.4f}")
print("\nConfusion matrix (test, normalisé):")
display(pd.DataFrame(confusion_matrix(Yt, ypred_test_n),
                     index=['true_0','true_1'], columns=['pred_0','pred_1']))
print("\nClassification report (test, normalisé):")
print(classification_report(Yt, ypred_test_n))

# 8) Résumé final et recommandations imprimées
print("====== Résumé rapide ======")
print(f"- Taille dataset: {df.shape[0]} échantillons, {len(features)} features.")
print(f"- Répartition binaire (0=bad,1=good):\n{df['quality_bin'].value_counts().to_dict()}")
print(f"- Meilleur k sans normalisation: {k_star} (err val={error_val[ind_opt]:.4f}, err test={error_test:.4f})")
print(f"- Meilleur k avec normalisation: {k_star_n} (err val={error_val_n[ind_opt_n]:.4f}, err test={error_test_n:.4f})")
print("\nRemarques:")
print("- Le scaling (StandardScaler) est important pour k-NN car k-NN utilise la distance Euclidienne.")
print("- Pour réduire la variance due au 'data split', utiliser validation croisée (cross-validation) ou répéter plusieurs splits aléatoires et moyenniser.")
print("- On peut aussi tester d'autres classifieurs (logistic regression, SVM, random forest) pour comparer.")

# (Optionnel) 9) Exemple rapide de cross-validation pour choisir k (si vous voulez)
from sklearn.model_selection import cross_val_score
print("\n(Optionnel) CV rapide pour k= {}, 5-fold sur l'ensemble d'entraînement normalisé".format(k_star_n))
clf_cv = KNeighborsClassifier(n_neighbors=k_star_n)
scores = cross_val_score(clf_cv, Xa_n, Ya, cv=5, scoring='accuracy')
print("CV accuracies:", scores)
print("CV mean accuracy:", scores.mean(), "std:", scores.std())
```
